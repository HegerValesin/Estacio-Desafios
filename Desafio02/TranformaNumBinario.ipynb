```python
# Universidade Estácio 
#Aluno: Heger Valesin - Matrícula: 202204474646
#Curso: Desenvolvedor FullStack - Desafio 02 - modulo

#função que vai converter o númerio decimala em binário.
#Confere sé o numero digitado é menor que 0
#Divide o numero digitado por 2 e vai acressentando o resto na variável binaria
#pega o resto e divide por 2 e arredonda para baixo
def decimal_binario(n_decimal):
      binario = '' 
      while n_decimal > 0:
        binario+= str(n_decimal%2) 
        n_decimal//= 2 
      return binario[::-1]


if __name__ == '__main__':
    numero = input("Digite um número: ")
    if not numero.isdigit():
      print("Digite apenas numeros!")
    else:
      print("O número em binário é: "+decimal_binario(int(numero)))
```
